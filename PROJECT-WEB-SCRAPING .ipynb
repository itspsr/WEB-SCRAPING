# python program to display all the header tags from wikipedia.org

```python
from bs4 import BeautifulSoup
import requests
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
soup = BeautifulSoup(page.content)
soup

header_tags = []

for i in soup.find_all('span',class_= "mw-headline"):

    header_tags.append (i.text)
    
header_tags
```

# python program to display IMDB’s Top rated movies

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page = requests.get('https://www.imdb.com/chart/top/')

page
soup = BeautifulSoup(page.content)

soup

years = []

for i in soup.find_all('span',class_="secondaryInfo"):
     years.append (i.text)
        
years[0:100]

titles = []

for i in soup.find_all('td',class_="titleColumn"):
    titles.append(i.text.replace('\n',''))
    
titles[0:100]

imdb_ratings = []

for i in soup.find_all('td', class_="ratingColumn imdbRating"):
    imdb_ratings.append(i.text.replace('\n',''))
    
imdb_ratings[0:100]

movies = pd.DataFrame({'Movies':titles[0:100],
                       'Years':years[0:100],
                       'Imdb_rating':imdb_ratings[0:100],})


movies
```

# Write a python program to scrape product name, price and discounts from https://meesho.com/bags-
