```python
href = []
lnks1=driver.find_elements_by_xpath("/html/body/ytd-app/div/ytd-page-manager/ytd-search/div[1]/ytd-two-column-search-results-renderer/div/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-video-renderer[2]/div[1]/div/div[1]/div/h3/a")
for lnk in lnks1:
    href.append(lnk.get_attribute('href'))
print(href)
```

```python
# empty list
comments = []
comment_upvote = []
Time = []

# auto scroll page
SCROLL_PAUSE_TIME = 2
CYCLES = 60

for i in href:
    driver.get(i)
    time.sleep(SCROLL_PAUSE_TIME)
    for i in range(CYCLES):
        html = driver.find_element_by_tag_name('html')
        driver.find_element_by_tag_name('html')
        
        time.sleep(SCROLL_PAUSE_TIME)
   
    
    # scraping the data
    try:
        cmnt = driver.find_elements_by_xpath("//*[@id='content-text']")
        for i in cmnt[0:500]:
            comments.append(i.text)
    except NoSuchElementException:
            comments.append('-')

    try:
        upvote = driver.find_elements_by_xpath("//*[@id='vote-count-middle']")
        for i in upvote[0:500]:
            comment_upvote.append(i.text)
    except NoSuchElementException:
            comment_upvote.append("-")
        

    try:
        time =  driver.find_elements_by_xpath("//*[@id='header-author']/yt-formatted-string/a")
        for i in time[0:500]:
            Time.append(i.text)
    except NoSuchElementException:
            Time.append('-')
```

```python
# creating data frame
df = pd.DataFrame({})
df['comments'] = comments
df['comment_upvote'] = comment_upvote
df['Time'] = Time
df
```
